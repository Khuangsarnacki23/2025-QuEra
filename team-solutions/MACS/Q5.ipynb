```python
import math
pi = math.pi
@move.vmove
def local_hadamard(state: move.core.AtomState, indices) -> move.core.AtomState:
    state = move.LocalXY(state, 0.25 * pi, 0.5 * pi, indices)
    state = move.LocalRz(state, pi, indices)
    state = move.LocalXY(state, -0.25 * pi, 0.5 * pi, indices)
    return state
```

```python
import math
@move.vmove
def problem5():
    q = move.NewQubitRegister(7)
    pi = math.pi
    state = move.Init(qubits=[q[0],q[1],q[2],q[3],q[4],q[5],q[6]], indices=[0,1,2,3,4,5,6])
    state.gate[[0,1,2,3,4,5]] = move.Move(state.storage[[0,1,2,3,4,5]])
    state = move.LocalXY(state,0,pi /2,[0,1,2,3,4,5])
    state.storage[[0]] = move.Move(state.gate[[0]])
    state.gate[[6]] = move.Move(state.storage[[6]])
    state = move.GlobalCZ(state)
    state = move.LocalXY(state,0,pi,[[6]])
    state.storage[[1]] = move.Move(state.gate[[1]])
    state.gate[[0]] = move.Move(state.storage[[0]])
    state = move.GlobalCZ(state)
    state.storage[[0]] = move.Move(state.gate[[0]])
    state = move.LocalXY(state,0,pi,[1,2,3,4,5,6]) 
    state.gate[[0,1]] = move.Move(state.storage[[0,1]])
    state.storage[[6]] = move.Move(state.gate[[6]]) 
    state.storage[[0,3,5,6]] = move.Move(state.gate[[0,3,5,6]])
    state = move.LocalXY(state,0,pi,[1,2,4])
    state.storage[[1,2,4]] = move.Move(state.storage[[1,2,4]])
    move.Execute(state)
    
```
